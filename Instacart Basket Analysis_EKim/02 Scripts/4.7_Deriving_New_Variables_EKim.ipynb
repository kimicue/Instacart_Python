# 4. Create a column “busiest_period_of_day” containing “Most orders,” “Average orders,” and “Fewest orders”

```python
# Display count of 'orders_day_of_week' column in 'ords_prods_merge' dataframe

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Display result_1 change using for-loop

result_1 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if value in [10,11,14,15,13,12,16,9,17]:
        result_1.append("most orders")
    elif value in [23,6,0,1,5,2,4,3]:
        result_1.append("fewest orders")
    else:
        result_1.append("average orders")
```

```python
# Creating new column 'busiest_period_of_day' and set it equal to result_1

ords_prods_merge['busiest_period_of_day'] = result_1
```
